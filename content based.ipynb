### removing the year from the title column and placing it in its own column

```python
#Using regular expressions to find a year stored between parentheses
#specify the parantheses so we don't conflict with movies that have years in their titles
movies_data['year'] = movies_data.title.str.extract('(\(\d\d\d\d\))',expand=False)
movies_data.head(3)
```

```python
#Removing the parentheses
movies_data['year'] = movies_data.year.str.extract('(\d\d\d\d)',expand=False)
movies_data.head(3)
```

```python
#Removing the years from the 'title' column
movies_data['title'] = movies_data.title.str.replace('(\(\d\d\d\d\))', '')
movies_data.head(3)
```

```python
#Applying the strip function to get rid of any ending whitespace characters that may have appeared
movies_data['title'] = movies_data['title'].apply(lambda x: x.strip())
movies_data.head()
```

```python
#Every genre is separated by a | so we simply have to call the split function on |
movies_data['genres'] = movies_data.genres.str.split('|')
movies_data.head()
```

# Onehot Encode

One-Hot-Encode the list of genres because it is categorical data. We store every different genre in columns that contain either 1 or 0. 1 shows that a movie has that genre and 0 shows that it doesn't. 

```python
# make a copy of the movies_data
movies_with_genres = movies_data.copy(deep=True)

# 1 if that column contains movies in the genre at the present index and 0 if not.

x = []
for index, row in movies_data.iterrows():
    x.append(index)
    for genre in row['genres']:
        movies_with_genres.at[index, genre] = 1

# Confirm that every row has been iterated and acted upon
print(len(x) == len(movies_data))

movies_with_genres.head(3)
```

```python
#Filling in the NaN values with 0 to show that a movie doesn't have that column's genre
movies_with_genres = movies_with_genres.fillna(0)
movies_with_genres.head(3)
```

# creating an input user to recommend movies to

```python
# so on a scale of 0 to 5, with 0 min and 5 max, see Deja's movie ratings below
deja_movie_ratings = [
            {'title':'Predator', 'rating':3},
            {'title':'Final Destination', 'rating':4.9},
            {'title':'Mission Impossible', 'rating':4},
            {'title':"Toy Story", 'rating':4.9},
            {'title':'Exorcist, The', 'rating':4.8},
            {'title':'Waiting to Exhale', 'rating':3.9},
            {'title':'Higher Learning', 'rating':4.5},
            {'title':'Omen, The', 'rating':2}
         ] 
deja_movie_ratings = pd.DataFrame(deja_movie_ratings)
deja_movie_ratings
```

```python
# Extracting movie Ids from movies_df and updating deja_movie_ratings with movie Ids.

deja_movie_Id = movies_data[movies_data['title'].isin(deja_movie_ratings['title'])]

# Merging Deja movie Id and ratings into the deja_movie_ratings data frame. 
# This action implicitly merges both data frames by the title column.

deja_movie_ratings = pd.merge(deja_movie_Id, deja_movie_ratings)

# Display the merged and updated data frame.

deja_movie_ratings
```

 filtering out the rows that contain the input movie's title and then merging this subset with the input dataframe.

```python
#Dropping information we don't need such as year and genres
deja_movie_ratings = deja_movie_ratings.drop(['genres','year'], 1)


#Final input dataframe
#If a movie you added in above isn't here, then it might not be in the original 
#dataframe or it might spelled differently, please check capitalisation.
deja_movie_ratings
```

```python
# filter the selection by outputing movies that exist in both deja_movie_ratings and movies_with_genres
deja_genres_df = movies_with_genres[movies_with_genres.movieId.isin(deja_movie_ratings.movieId)]
deja_genres_df
```

```python
# First, let's reset index to default and drop the existing index.
deja_genres_df.reset_index(drop=True, inplace=True)

# Next, let's drop redundant columns
deja_genres_df.drop(['movieId','title','genres','year'], axis=1, inplace=True)

# Let's view chamges

deja_genres_df
```

```python
# let's confirm the shapes of our data frames to guide us as we do matrix multiplication

print('Shape of Deja_movie_ratings is:',deja_movie_ratings.shape)
print('Shape of Deja_genres_df is:',deja_genres_df.shape)
```

```python
#finding the dot product of transpose of deja_genres_df by deja rating column
deja_profile = deja_genres_df.T.dot(deja_movie_ratings.rating)

deja_profile
```

Deja has a clear love for thrillers,dramas and comedys.

```python
#set the index to the movieId
movies_with_genres = movies_with_genres.set_index(movies_with_genres.movieId)

movies_with_genres.head()
```

```python
# Deleting four unnecessary columns.
movies_with_genres.drop(['movieId','title','genres','year'], axis=1, inplace=True)

movies_with_genres.head()
```

```python
# Multiply the genres by the weights and then take the weighted average.
recommendation_table_df = (movies_with_genres.dot(deja_profile)) / deja_profile.sum()

recommendation_table_df.head()
```

```python
# sort values ascending
recommendation_table_df.sort_values(ascending=False, inplace=True)

recommendation_table_df.head(20)
```

# Recommendations

```python
# make a copy of the original movies_data
copy = movies_data.copy(deep=True)

# set its index to movieId
copy = copy.set_index('movieId', drop=True)

# enlist the top 20 recommended movieIds we defined above
top_20_index = recommendation_table_df.index[:20].tolist()

# slice these indices from the copied movies df and save in a variable
recommended_movies = copy.loc[top_20_index, :]

recommended_movies
```
